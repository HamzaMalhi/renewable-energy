```python
# Function for reading file:
def read_windSolar_data():
    

    # Remove unnecessary columns
    df = df.drop(columns=['quantity_footnotes', 'category', 'unit', 'quantity', 'commodity_transaction'])

    # Pivot the dataframe to have years as columns
    df_years = df.pivot(index='country_or_area', columns='Year')

    # Pivot the dataframe to have countries as columns
    df_countries = df.pivot(index='Year', columns='country_or_area')

    # Clean the dataframes
    df_years = df_years.dropna()
    df_countries = df_countries.dropna()

    return df_years, df_countries
```

```python
# Create a dataframe with years as columns
df_years = df.pivot_table(index='country_or_area', columns='Year', aggfunc='first')
# Create a dataframe with countries as columns
df_countries = df.pivot_table(index='Year', columns='country_or_area', aggfunc='first')
```
